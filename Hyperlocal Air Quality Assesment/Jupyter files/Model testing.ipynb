```python
# Adjacency matrix
import random
no_of_hours=24
def BP(land_use,w_land,traffic):
    land_use_data = land_use    # 22 x 6
    weights_landuse = w_land    # 1 x 6
    BP_row = np.dot(weights_landuse, land_use_data.T)   # 1 x 22
    BP_static = np.array([BP_row for t in range(24)])   # 24 x 22

    no_of_vehicles = [] # 24 x 22 # might depend on width of the road, points of interest
    weights_vehicle =  0 # 1 x 1
    BP_dynamic = weights_vehicle*np.array(no_of_vehicles)  # 24 x 22
    
    BP_final = np.add(BP_static+BP_dynamic)
    return BP_final

def pol(land_use,w_land,traffic,Kij):
    BP_final = BP(land_use,w_land,traffic)
    ad=[None]*22
    ad[0]=[2,22]
    ad[1]=[1,3]
    ad[2]=[2,4,8]
    ad[3]=[3,5]
    ad[4]=[4,6]
    ad[5]=[5,7,9]
    ad[6]=[6,8]
    ad[7]=[3,7,12,21]
    ad[8]=[6,10,11]
    ad[9]=[9,13]
    ad[10]=[9,12]
    ad[11]=[8,11,13]
    ad[12]=[10,12,14,18]
    ad[13]=[13,15]
    ad[14]=[14,16]
    ad[15]=[15,17]
    ad[16]=[16,19]
    ad[17]=[13,19]
    ad[18]=[17,18,20]
    ad[19]=[19,22]
    ad[20]=[8,22]
    ad[21]=[1,20,21]

    # Distance matrix
    dis=[None]*22
    dis[0]=[2,3]
    dis[1]=[2,4]
    dis[2]=[4,3,4]
    dis[3]=[3,2]
    dis[4]=[2,3]
    dis[5]=[3,2,5]
    dis[6]=[2,2]
    dis[7]=[4,2,4,3]
    dis[8]=[5,3,3]
    dis[9]=[3,3]
    dis[10]=[3,3]
    dis[11]=[4,3,7]
    dis[12]=[3,7,5,4]
    dis[13]=[5,2]
    dis[14]=[2,3]
    dis[15]=[3,3]
    dis[16]=[3,6]
    dis[17]=[4,5]
    dis[18]=[6,5,8]
    dis[19]=[8,7]
    dis[20]=[3,3]
    dis[21]=[3,7,3]

    dis=[[250*j for j in dis[i]] for i in range(22)]

    # BP of each node
        # for now choosing randomly, later on BP is determined by several other factors
        # BP is list of list which contains Background polution data of all 22 nodes for all 24 hours

    BP = BP_final
    #print(BP)
    # Convective transfer
        # convective tansfer coefficient (K)
        # for now choosing K to be a constant, later on it might be varying for different nodal interactions
    K = [] 
    for t in range(no_of_hours):
        k=0
        out_1=[]
        for i in range(22):
            out=[]
            for j in range(22):
                if j+1 in ad[i]:
                    out.append(Kij[t][k])
                    k+=1
                else:
                    out.append(0)
            out_1.append(out)
        #print(out_1)
        K.append(out_1)
    #print(K)        

    # pollution data for each node
    pol=[[0 for i in range(22)] for j in range(no_of_hours)]
    pol_i_to_j=[[[0 for i in range(22)] for j in range(22)] for k in range(no_of_hours)]
    pol_j_to_i=[[[0 for i in range(22)] for j in range(22)] for k in range(no_of_hours)]

    # starting case at t=0
    pol[0]=df_2.values[0].tolist()
    pol_i_to_j[0]=[[0]*22]*22
    pol_j_to_i[0]=[[0]*22]*22

    # convective transfer
    dis_inv=[[1/i for i in dis[j]] for j in range(22)]


    # at any time t
    for t in range(1,no_of_hours):

        # convective transfer
        # i to j
        for i in range(22):
            for j in ad[i]:
                pol_i_to_j[t][i][j-1]=K[t][i][j-1]*pol[t-1][i]*dis_inv[i][ad[i].index(j)]/sum(dis_inv[i])

        # j to i
        for i in range(22):
            for j in ad[i]:
                pol_j_to_i[t][i][j-1]=K[t][i][j-1]*pol[t-1][j-1]*dis_inv[i][ad[i].index(j)]/sum(dis_inv[j-1])

        # pollution concentration
        pol[t]=[(pol[t-1][i]+BP[t][i]-sum(pol_i_to_j[t][i])+sum(pol_j_to_i[t][i])) for i in range(22)]
    
    return pol # 24 x 22
```

```python
# RMSE Computation
from sklearn.metrics import mean_squared_error as mse
import pandas as pd
import numpy as np

def cost(df_1,df_2):
    # computing error for each columns
    error = []
    n = len(df_1.values[0])

    if len(df_1.values[0])==len(df_2.values[0]):
        for i in range(n):
            x = df_1.iloc[:,i]
            y = df_2.iloc[:,i]
            error.append(mse(x,y,squared=False))

    # final rmse (subjected to change)
    rmse = np.array(error).mean()
    model_rmse = rmse/22
    return model_rmse
```
